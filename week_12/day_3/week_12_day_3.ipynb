## Question 2

Join the data containing card details and customer details by customer id, so that all records of card details and any matching records in customer details are kept.

```python
card_base.merge(cust_base, how = 'left', on = 'cust_id')
```

## Question 3

Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
fraud.merge(trans_base, how = 'outer', on = 'transaction_id')
```

Since the dataframe after the merge has 10000 rows, like the transaction_base original df, means our data set is complete and all fraud details were previously contained in the transaction_base.

## Question 4

Join the data containing card details with transaction details so rows from the first which have matching ones in the second are returned.



```python
card_base.merge(trans_base, how = 'left', left_on = 'card_number', right_on = 'credit_card_id' )
```

Read in hat_observations and split observation into two columns, hat_colour and hat_type.

## Question 6

Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where the most berets were observed.

Hints:

- you can use the .agg() function to combine columns
- you may need to convert the data types of the columns you wish to combine

```python
cols=['day','month','year']
hat_obs['date'] = hat_obs[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
```
